```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def convert(text):
    L=[]
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
movies['genres']=movies['genres'].apply(convert)

movies
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
#for cast column
def convert2(text):
    counter=0
    L=[]
    for i in ast.literal_eval(text):
        if counter<3:
            L.append(i['name'])
            counter+=1
    return L
            
```

```python
movies['cast']=movies['cast'].apply(convert2)
```

```python
#for crew column
def fetch_director(text):
    L=[]
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            L.append(i['name'])
    return L
        
```

```python
movies['crew'] = movies['crew'].fillna('[]')
```

```python
movies = movies[movies['crew'].notnull()]
movies = movies[movies['crew'] != '[]']
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
movies
```

```python
##to remove spaces
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
    
```

```python
movies['cast'] = movies['cast'].apply(collapse)
movies['crew'] = movies['crew'].apply(collapse)
movies['genres'] = movies['genres'].apply(collapse)
movies['keywords'] = movies['keywords'].apply(collapse)
```

```python
movies.sample(5)
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(5)
```

```python
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
new = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new['tags']=new['tags'].apply(lambda x:" ".join(x))
```

```python
new['tags']=new['tags'].apply(lambda x:x.lower())
```

```python
new
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector=cv.fit_transform(new['tags']).toarray()
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(vector)
```

```python
#  is used to retrieve the index of the first row in a DataFrame new where
#the value in the title column matches.
c=new[new['title']=='The Avengers'].index[0]
c
```

```python
def recommend(movie):
    index=new[new['title']==movie].index[0]
    distances=sorted(list(enumerate(similarity[index])),reverse=True,key=lambda x:x[1])
    for i in distances[1:6]:
        print(new.iloc[i[0]].title)
    
```

```python
recommend('Superman')
```
